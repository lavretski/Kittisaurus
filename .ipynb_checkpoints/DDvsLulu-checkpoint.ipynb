```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory
import matplotlib.pyplot as plt
def dd_lulu(inp):
    if inp == 0:
        return 'dd'
    return 'lulu'
```

```python
vgg16_net = tf.keras.applications.VGG16(weights = 'imagenet', include_top = False, 
                                       input_shape = (150, 150, 3))
```

```python
vgg16_net.trainable = False
```

```python
vgg16_net.summary()
```

```python
model = tf.keras.models.Sequential([
  vgg16_net,
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(256, activation='relu'),
  tf.keras.layers.Dropout(0.5),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=tf.keras.optimizers.Adam(learning_rate = 1e-5), 
              metrics=['accuracy'])
```

```python
train_dataset = image_dataset_from_directory('dd_lulu_data_fromgooglephotos/train',
                                             batch_size = 5,
                                             image_size=(150, 150),
                                             label_mode = 'binary',
                                            )
```

```python
test_dataset = image_dataset_from_directory('dd_lulu_data_fromgooglephotos/test',
                                             batch_size = 1,
                                             image_size=(150, 150),
                                             label_mode = 'binary',
                                            )
```

```python
with tf.device('/GPU:1'):
    model.fit(train_dataset, epochs = 5)

#model.fit(train_dataset, epochs = 5) #with cpu
```

```python
model.evaluate(train_dataset)
```

```python
for item in test_dataset:
        plt.imshow(item[0][0] / 255.0)
        plt.show()
        a = (model.predict(item[0]) > 0.5)[0]
        print('nn answer', dd_lulu(a))
        input()
```
