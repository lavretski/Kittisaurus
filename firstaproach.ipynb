```python
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Conv2D, MaxPooling2D
from tensorflow.python.keras.layers import Activation, Dropout, Flatten, Dense
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image_dataset_from_directory
import tensorflow as tf
import numpy as np
def dd_lulu(inp):
    if inp == 0:
        return 'DD'
    return "Lulu"
```

```python
train_dataset = image_dataset_from_directory('train',
                                             batch_size = 5,
                                             image_size=(150, 150),
                                             label_mode = 'binary',
                                            )
```

```python
test_dataset = image_dataset_from_directory('test',
                                             batch_size = 1,
                                             image_size=(150, 150),
                                             label_mode = 'binary',
                                            )
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Conv2D(20, (5, 5), activation = 'relu', input_shape = (150, 150, 3)),
  tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='valid'),
  tf.keras.layers.Conv2D(64, (5, 5), activation = 'relu', input_shape = (28, 28, 1)),
  tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2), padding='valid'),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(100, activation='relu'),
  tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer="adam",
              loss = tf.keras.losses.BinaryCrossentropy(from_logits=False),
              metrics = ['accuracy'])
```

```python
model.fit(train_dataset, epochs = 1)
```

```python
model.evaluate(test_dataset)
```

```python
for item in test_dataset:
        plt.imshow(item[0][0] / 255.0)
        plt.show()
        print('correct answer', dd_lulu(bool(item[1])))
        a = (model.predict(item[0]) > 0.5)[0]
        print('nn answer', dd_lulu(a))
        input()
```
